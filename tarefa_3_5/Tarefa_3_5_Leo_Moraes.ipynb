#### b-) Ler o dataset "*Hourly_wages/hourly_wages.csv*" como dataframe utilizando a biblioteca Pandas e mostrar as 5 primeiras e 5 últimas linhas.

```
# import libs
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


df = pd.read_csv('/content/DSBD/Datasets/Hourly_wages/hourly_wages.csv')

# return
df
```

#### c-) Mostrar os histogramas dos atributos: *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
# Setando cada variável

fig, axes = plt.subplots(1, 3, figsize = (15, 5), sharey = True)
fig.suptitle('Histogramas', fontsize = 20)


# education_yrs
sns.histplot(ax=axes[0], data = df, x='education_yrs', kde = True)

# experience_yrs
sns.histplot(ax=axes[1], data = df, x='experience_yrs', kde = True)

# experience_yrs
sns.histplot(ax=axes[2], data = df, x='age', kde = True)

plt.show()
```

#### d-) Mostrar os gráficos de setores (*pie chart*) de cada um dos atributos: *union*, *female*, *marr*, *south*, *manufacturing* e *construction* num mesmo quadro de plotagem (usando subplots).

```
# obtendo a proporção  de resultados

union = df['union'].value_counts()
female = df['female'].value_counts()
marr = df['marr'].value_counts()
south = df['south'].value_counts()
manufacturing = df['manufacturing'].value_counts()
construction = df['construction'].value_counts()


# Fazendo o gráfico em pizza
sns.set()
fig, axes = plt.subplots(2, 3, figsize = (15, 10))
fig.suptitle('Gráficos de setpres', fontsize = 20)

# female
axes[0, 0].pie([union[0], union[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[0, 0].set_title('union')

# marr
axes[0, 1].pie([female[0], female[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[0, 1].set_title('female')

# south
axes[0, 2].pie([south[0], south[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[0, 2].set_title('south')

#south
axes[1, 0].pie([marr[0], marr[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[1, 0].set_title('marr')

# manufeactuing
axes[1, 1].pie([manufacturing[0], manufacturing[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[1, 1].set_title('manufacturing')

# construction
axes[1, 2].pie([construction[0], construction[1]], labels = ['0', '1'], autopct = '%1.f', shadow = True)
axes[1, 2].set_title('construction')


plt.show()
```

#### e-) Mostrar os boxplots das variáveis *education_yrs*, *experience_yrs* e *age* num mesmo quadro de plotagem (usando *subplots*).

```
# Setando cada variável
sns.set()
sns.set_palette('Set1', 8, .75)
fig, axes = plt.subplots(1, 3, figsize = (15, 5), sharey = True)
fig.suptitle('Boxplots', fontsize = 20)


# education_yrs
sns.boxplot(ax=axes[0], data = df, x='education_yrs')

# experience_yrs
sns.boxplot(ax=axes[1], data = df, x='experience_yrs')

# experience_yrs
sns.boxplot(ax=axes[2], data = df, x='age')

plt.show()
```

#### f-) Mostre lado a lado os boxplots da variável *age* discriminados por sexo (female).

```
plt.figure(figsize=(10, 6))
sns.boxplot(data = df, x="female", y="age")
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

#### g-) Responda observando os boxplots construídos no item anterior:

ii-) Qual é aproximadamente o 2o. quartil das idades das mulheres? Resposta: 35 anos

iii-) Qual é aproximadamente o 1o. quartil das idades dos homens? Resposta: 27 anos



#### h-) Mostre lado a lado os boxplots da variável *wage_per_hour* discriminados por sexo (female).

```
plt.figure(figsize=(10, 6))
sns.boxplot(data = df, x="female", y="wage_per_hour")

plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.xlabel('Female Indicator')
plt.ylabel('Wage per Hour')
plt.show()
```

#### j-) Mostre lado a lado os boxplots da variável wage_per_hour discriminados por estado civil (marr).

```
plt.figure(figsize=(10, 6))
sns.boxplot(data = df, x="marr", y="wage_per_hour")

plt.xticks(ticks=[0, 1], labels=['Not Married', 'Married'])
plt.xlabel('Marital Status')
plt.ylabel('Wage per Hour')
plt.show()
```
